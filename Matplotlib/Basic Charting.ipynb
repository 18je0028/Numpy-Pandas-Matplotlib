```python
# Add title and labels for x & y axis
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x,y)

plt.title('Our First Graph!')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
# Change font type, size, etc.
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x,y)

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
# Change tick marks (scale graph)
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1, 2, 3, 4])
plt.yticks([2, 4, 6, 7, 7.5, 10])

plt.show()
```

```python
# Add a legend
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.plot(x, y, label = '2x')

# Due to this line legend will be shown when we do (plt.show())
plt.legend()

plt.show()
```

```python
# Restyle our line (color, line style, markers, width)
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label = '2x', color = 'red', linewidth = 2 , marker = '.', markersize = 10, markeredgecolor = 'green')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
# Restyle our line (color, line style, markers, width)
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label = '2x', color = 'red', linewidth = 2 , marker = '.', linestyle = '--',
         markersize = 10, markeredgecolor = 'green')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
# Shorthand notation to restyle lines
# Restyle our line (color, line style, markers, width)
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, 'r.--', label = '2x')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
# Shorthand notation to restyle lines
# Restyle our line (color, line style, markers, width)
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, 'b^--', label = '2x')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
#  Plot more complex lines

x = np.arange(0,4.5,0.5)
plt.plot(x, x**2, 'r.--', label = 'x^2')

plt.title('Our Second Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
#  Plot more complex lines
x = np.arange(0,4.5,0.5)
plt.plot(x, x**2, label = 'x^2')

plt.title('Our Second Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
# Resize Graph
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

# resize code
plt.figure( figsize=(5,3), dpi = 120)

plt.plot(x, y, 'b^--', label = '2x')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
# Save graph
x = [1, 2, 3, 4 , 5]
y = [2, 4, 6, 8, 10]

# resize code
plt.figure( figsize=(5,3))

plt.plot(x, y, 'b^--', label = '2x')

plt.title('Our First Graph!', fontdict = {'fontname' : 'comic sans MS', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.savefig('myGraph.png', dpi = 300)

plt.show()
```

```python
#  Bar Chart
genders = ['Men', 'Women', 'Other']
population = [55, 44, 1]

bars = plt.bar(genders,population)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()               
```

```python
#  Bar Chart
genders = ['Men', 'Women', 'Other']
population = [55, 44, 1]

bars = plt.bar(genders,population)

patterns = ['/', 'o', '*']

for i in range(0,3):
    bars[i].set_hatch(patterns[i])

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.show()           
```
